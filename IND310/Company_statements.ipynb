```python
import pandas as pd
#url="https://raw.githubusercontent.com/jfrausig/test/master/ticker_3.csv"
#df=pd.read_csv(url, error_bad_lines=False)
df=pd.read_csv("http://pvexpect.com/IND310_2021_ESG/portfolio_3.csv", delimiter=";")

aktier=[]
for index, row in df.iterrows():
    d=row.to_dict()
    aktier.append(d)
    #print(d)
aktier=pd.DataFrame(aktier)

aktier
```

```python
aktier = aktier.assign(**{'aktie': ""})

aktier.index=aktier.ticker
for i, ticker in enumerate(aktier[aktier['Virksomhed']. notna()].ticker):

  name=aktier['Virksomhed'][ticker]
  aktier['Virksomhed'][ticker]=str(name).replace(".","").replace("-","").replace("(","").replace(")","").replace("/","")

aktier #[aktier['shortName']. notna()]
```

```python
import pandas as pd
import investpy
import pandas as pd
financials=pd.DataFrame()

#valgte_virksomheder=virksomheds_database

for i in range(0,len(aktier)):

  stock=aktier["ticker"][i].replace(".HE","").replace(".ST","").replace(".CO","").replace(".DE","").replace("-","").replace(".AX","").replace(" ","")
  country=aktier["country"][i]
  Virksomhed=aktier["Virksomhed"][i]
  print(stock)
  try:
    income_statement_annual=investpy.stocks.get_stock_financial_summary(stock, country, summary_type='income_statement', period='annual') #balance_sheet cash_flow_statement  balance_sheet
    
    income_statement_quarterly=investpy.stocks.get_stock_financial_summary(stock, country, summary_type='income_statement', period='quarterly') #*4 #balance_sheet cash_flow_statement  balance_sheet
    
    income_statement_quarterly['Virksomhed']=Virksomhed
    income_statement_quarterly['period']="quarterly"
    #income_statement_quarterly=income_statement_quarterly.add_suffix("_qrt")
    income_statement_quarterly['Datetime']=income_statement_quarterly.index
    income_statement_annual['Virksomhed']=Virksomhed
    income_statement_annual['period']="annual"
    income_statement_annual['Datetime']=income_statement_annual.index
    financials=pd.concat([financials,income_statement_annual,income_statement_quarterly], axis=0)
    #print(financials)
  except:
    print('none')
```

```python
import altair as alt
from vega_datasets import data
import pandas as pd
"""
source = data.anscombe().copy()
source['line-label'] = 'x=y'
source = pd.concat([source,source.groupby('Series').agg(x_diff=('X','diff'), y_diff=('Y','diff'))],axis=1)
source['rate'] = source.y_diff/source.x_diff
source['rate-label'] = 'line y=x'
"""

Virksomhed=alt.binding_select(options=list(financials.Virksomhed.unique()))
periodicity=alt.binding_radio(options=['quarterly','annual'])

scatter = alt.Chart(financials).transform_fold(list(financials.drop(['Datetime','period','Virksomhed'], axis=1)), as_=['key_y', 'value_y']).mark_bar(size=20, opacity=0.60).encode(
    x='Datetime:T',
    y=alt.Y('value_y:Q', stack=None),
    color='key_y:N',
    opacity= alt.condition(
        alt.datum.period == "quarterly", 
        alt.value(1),
        alt.value(0.7))

    #tooltip=['X','Y','rate']
)

regression=  scatter.transform_filter(
    'datum.period==periodicity.periodicity'
).transform_regression('Datetime', 'value_y',
        groupby=['key_y']).transform_fold(
     ["reg-line"], 
     as_=["Regression", "y"]
).mark_line(
      size=5,
     opacity=0.50, 
     strokeDash=[3,5],
     shape='mark'
).encode(alt.Color("Virksomhed:N"))




text =  scatter.transform_filter(
    'datum.period==periodicity.periodicity'
).transform_regression('Datetime', 'value_y',
        groupby=['key_y'], params=True).mark_text(align='left').encode(
    x=alt.value(20),
    y=alt.value(20),
    #y=alt.Y('key_y:O',axis=None),
    #x=alt.value(20),  # pixels from left
    
    #y=alt.Y('cor:N',axis=None), #alt.value(30),  # pixels from top
    text='cor:N',
    #color="stock:N", #
    #tooltip='key_y:N',
).transform_calculate(
  #period_select='toString(quarterly)',
  cor_unformat='sqrt(datum.rSquared)',
  cor="(' b: '+format(datum.coef[1]*365*24*60*100, '.3f')+' Cor:'+format(datum.cor_unformat, ',.3f')+',  R2: '+format(datum.rSquared, ',.5f'))", #+'  ,a: '+format(datum.coef[0], '.1f')
)


numeric_text = scatter.mark_text(
    align='center',
    baseline='middle',
    dy=-10,  # Nudges text up so it doesn't appear on top of the bar
    color="darkred"
).encode(
    text="value_y:Q",

)



scatter = scatter +regression +text +numeric_text


chart = scatter.facet(
    columns=2,
    facet=alt.Facet('key_y:N',header=alt.Header(labelFontSize=25))
).resolve_scale(
    x='independent',
    y='independent'
).add_selection( #tilføj dropdown menu +graenesvaerdi 
    #alt.selection_single(fields=['virksomhed'], bind=virksomhed , name='virksomhed', init={"virksomhed":financials.stock[0]}),
    alt.selection_single(fields=['periodicity'], bind=periodicity , name='periodicity', init={"periodicity":"annual"})
).add_selection( #tilføj dropdown menu +graenesvaerdi 
    alt.selection_single(fields=['Virksomhed'], bind=Virksomhed , name='Virksomhed', init={"Virksomhed":financials.Virksomhed[0]}),
).transform_filter(
    'datum.Virksomhed==Virksomhed.Virksomhed'
)


#chart
chart.save("regnskaper.html")
```
