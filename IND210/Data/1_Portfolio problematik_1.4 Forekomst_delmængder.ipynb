<a name="Hent_bibliotekker_med_verktøyer"></a>
###**Hent bibliotekker av verktøyer** - `Import`

Python-kode i en modul får tilgang til koden i en annen modul ved å importere den. 
I Python er en modul ganske en fil i .py-filformat som inneholder Python-kode.
Det mest populære bibliotek til data analyse er pandas. 
Pandas er en data struktur, litt likesom en tabel i Excel med rekker og kolonner.

Import funksjonen kombinerer to operasjoner; den søker etter den navngitte modulen - `pandas` - , så binder den modulet til et navn, i dette eksempel `pd`.
Vanligvis foretrekker vi forekortelser, og importere bibliotekker, i forkortet form: `pandas as pd`. Fremover hetter bibliotekket `pd`

```
import pandas as pd
```

Genereltset importer vi funksjoner såledelse
```
import vårfunksjon
```

Alternativet til å importere funksjoner er å definere funksjonen direkte: 
```
def vårfunksjon():
  return x + 2
```

<a name="konvertere_til_pdf"></a>

##**Konvertere til pdf fil** - `tabula.convert_into`



Bruk Tabula for å gjøre PDF-rapporter til Excel-regneark, CSV og JSON-filer for bruk i analyse- og databaseapplikasjoner.

Konverter tabeller fra PDF til en fil. Utdatafilen lagres i output_path.

`tabula.io.convert_into(input_path, output_path, output_format='csv', stream='true',pages='all')`


| Parameter    | Format                       | Beskrivelse                                                                                                                     |
|---------------|------------------------------|---------------------------------------------------------------------------------------------------------------------------------|
| input_path    | Fil som objekt               | PDF-fil som skal konverteres                                                                                                    |
| output_path   | string                       | Filsti til ny datafil                                                                                                           |
| output_format | string                       | Dataformat på ny fil (csv, json eller tsv: Standard: csv)                                                                       |
| stream        | boolean (sannt /falsk)       | Tving PDF til å data trekkes ut (hvis det ikke er linjer som skiller hver celle, som i en PDF i et Excel-regneark)              |
| pages         | str, `int`, liste over: int. | En valgfri verdi som spesifiserer sider å trekke ut fra. Standard: 1. Hvis du vil trekke ut alle sidene, angir du sider = "alle". |




Denne modulen er en bibliotek innpakning eller på engelsk "wrapper" av tabula, som muliggjør utvinning av tabeller fra en PDF. 



En "wrapper" er en designmønster som brukes når du arbeider med relativt kompliserte funksjoner. Dette lar deg løse oppgaver generelt uten å vite detaljene.
Innpakningsfunksjonen utfører vanligvis noen håndteringsarbeide som
* tildeling og disponering av ressurser
* sjekke forholdene før og etter
* caching / resirkulering et resultat av en langsom beregning

En wrapper er fullt kompatibel med den innpakket funksjonen, slik at den kan brukes i stedet for den orignale funksjon.


Denne modulen trekker ut tabeller fra en PDF til en pandas DataFrame. Foreløpig bruker implementeringen av denne modulen underprosess.

<a name="Omnavngiv"></a>

##**Omnavngiv** - `as`

Du kan gi navn til modulfilen hva du vil, men den må ha filtypen .py

Du kan opprette et alias når du importerer en modul, ved å bruke som nøkkelord: `as`

For eksempel, et bibliotek til å vise grafer er `matplotlib`. Bibliotekket har et langt navn og benytter et undermodul. Derfor vil det være krevene alltid å skrive det lange navn.
En kjennt omnavngivning er 

```
import matplotlib.pyplot as plt
```







Importere under modul i matplotlib med kort navngivning `plt`

````
import matplotlib.pyplot as plt
````

Vis en graf uten å skrive hele navnet `matplotlib.pyplot`

````
plt.plot([1,2,3,4])
plt.ylabel('some numbers')
plt.show()
````
